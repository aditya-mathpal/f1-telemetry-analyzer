```python
import fastf1
from fastf1 import plotting
from fastf1.mvapi import get_circuit_info
import matplotlib.pyplot as plt
```

```python
import os
os.makedirs('cache', exist_ok=True)  
fastf1.Cache.enable_cache('cache')
```

```python
session = fastf1.get_session(2025, 'Silverstone', 'R')
session.load()
session.results
```

```python
session.laps
```

```python
ver_lap = session.laps.pick_drivers('VER').pick_fastest()
nor_lap = session.laps.pick_drivers('NOR').pick_fastest()
ham_lap = session.laps.pick_drivers('HAM').pick_fastest()
hul_lap = session.laps.pick_drivers('HUL').pick_fastest()
```

```python
ver_tel = ver_lap.get_car_data().add_distance()
nor_tel = nor_lap.get_car_data().add_distance()
ham_tel = ham_lap.get_car_data().add_distance()
hul_tel = hul_lap.get_car_data().add_distance()
```

```python
# plt.plot(ver_tel['Distance'], ver_tel['Speed'], label = 'Verstappen')
# plt.plot(nor_tel['Distance'], nor_tel['Speed'], label = 'Norris')
plt.plot(ham_tel['Distance'], ham_tel['Speed'], label = 'Hamilton', alpha=0.7)
plt.plot(hul_tel['Distance'], hul_tel['Speed'], label = 'Hulkenburg', alpha=0.7)
plt.xlabel('Distance (m)')
plt.ylabel('Speed (kmph)')
plt.title('Speed over Distance - Grand Prix (Silverstone 2023)')
plt.legend()
plt.show()
```
